```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt 

N = norm.cdf

def bas_call_value(x, c, T, r, sigma):
    d1 = (np.log(x/c) + (r + sigma**2/2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return x * N(d1) - c * np.exp(-r*T)* N(d2)
```

```python
## Values 
# x - Current Asset Price
# c - Strike Price 
# T - time to maturity
# r - risk-free rate
# sigma - annualized volatility of the asset's returns 

c = 100
r = 0.1
T = 2
sigma = 0.3
x = np.arange(60,140,0.1)
```

```python
# Use the model to predict the call values 
calls = [bas_call_value(s, c, T, r, sigma) for s in x]
```

```python
plt.plot(x, calls, label='Call Value')
plt.xlabel('$X_0$')
plt.ylabel('Call Option Value')
plt.show()
```

```python
# Analyse the plot according to the research paper in reference to the max value (stock price)
# and min value (stock price - exercise price) and one more plot showing how time affects the call values
calls2 = [bas_call_value(s, c, T/2, r, sigma) for s in x]
max_call_value = x
min_call_value = x - c*(np.ones(len(x)))
min_call_value[min_call_value < 0] = 0
print(max_call_value)
print(min_call_value)
```

```python
# Plot the values with reference values of max and min across stock prices for various times
plt.plot(x, calls, label='Call Value 2 year')
plt.plot(x, calls2, label='Call Value 1 year')
plt.plot(x, max_call_value, label='Call Value infinte years')
plt.plot(x, min_call_value, label='Call Value almost zero years')
plt.xlabel('$X_0$')
plt.ylabel('Call Option Value')
plt.legend()
plt.show()
```

```python
# plot call values for a given current stock price with respect to time
T_range = np.arange(0.00001, 2, 0.01)
S = 100
calls = [bas_call_value(S, c, t, r, sigma) for t in T_range]
plt.plot(T_range, calls, label='Call Value')
plt.xlabel('$T$ in years')
plt.ylabel(' Value')
plt.legend()
plt.show()
```

```python
# Plot and check around what time it shows the tendency to move towards 100 
T_range = np.arange(0.00001, 100, 0.01)
S = 100
calls = [bas_call_value(S, c, t, r, sigma) for t in T_range]
plt.plot(T_range, calls, label='Call Value')
plt.xlabel('$T$ in years')
plt.ylabel(' Value')
plt.legend()
plt.show()
```

```python
# Effect of various volatality of stocks i.e. sigma 
Sigmas = np.arange(0.00001, 1.5, 0.01)
calls = [bas_call_value(100, 100, 1, r, sig) for sig in Sigmas]
plt.plot(Sigmas, calls, label='Call Value')
plt.xlabel('$\sigma$')
plt.ylabel(' Value')
plt.legend()
plt.show()
```
